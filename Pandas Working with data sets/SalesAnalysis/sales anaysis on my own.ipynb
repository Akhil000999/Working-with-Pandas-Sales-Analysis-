```python
# creating aa empty data frame 

df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")
merge_df = pd.DataFrame()

files= [a for a in os.listdir("./Sales_Data")]

# merging all the files together as the one file 
for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    merge_df = pd.concat([merge_df,df])

#printing the all sales analysis merged file
#merge_df.head()

merge_df.to_csv("Modified1.csv",index=False)
```

```python
df1 = pd.read_csv("Modified1.csv")

df1.head()
```

```python
# daleting the NAN values from the data set

df1.drop([1],inplace=True)
df1.head()
```

```python
# reshaping the index values in the data frame 
df1.reset_index(drop=True,inplace=True)
```

```python
df1.head(10)
```

### before performing we want to do data cleaning 

```python
df1['Month']=df1["Order Date"].str[0:2]
df1.head()
```

```python
# how many NAN values are there in the column

df1.isnull().sum()
```

```python
# how to get all the row's containg NAN values 

nan_df= df1[df1.isnull().any(axis=1)]
nan_df.head()
```

```python
# DROPPING ALL THE "NAN VALUES FROM ROWS"
df1 = df1.dropna(how="all")
```

```python
df1.isnull().sum()
```

```python
# Finding  the values containing "or" 
df1=df1[df1["Month"].str[0:2]!="Or"]
df1.head()
```

```python

df1["Month"]= df1['Month'].astype('int32')
df1.head()
```

```python
df11= df1

df11
```

```python
# again we want to clean up the data str to int

df11['Quantity Ordered']=pd.to_numeric(df11['Quantity Ordered'])
df11['Price Each']=pd.to_numeric(df11['Price Each'])
df11.head()
```

### WHAT IS HIGHEST SALES MONTH

```python
# LET'S create anthor column for knowing which month having the highest sales 

df11['highest_sales']= df11['Quantity Ordered']*df11['Price Each']
df11.head()
```

#### which month is the highest earned month or which season


```python
results=df11.groupby('Month').sum()
results
```

### plotting the graph

```python
months= range(1,13,1)
plt.title("sales anaysis for a particular month's",fontdict={"fontsize":20})
plt.ylabel("Amount in the american doller's($)")
plt.xlabel('Months')
plt.xticks(months)
plt.bar(months,results['highest_sales'],label="sales analysis")
plt.legend()
plt.show()

plt.title("sales anaysis for a particular month's",fontdict={"fontsize":20})
plt.ylabel("Amount in the american doller's($)")
plt.xlabel('Months')
plt.plot(months,results['highest_sales'],label="sales analysis")
plt.xticks(months)
plt.legend()
plt.show()
```
