### 2) what is the city have highest number of the sales

```python
# we want to create a city column for knowing the which city having the highest sales

def grab(address):
    return address.split(',')[1]

def grab_state(address):
    return address.split(',')[2].split(' ')[1]
df1["City"]= df1['Purchase Address'].apply(lambda x: grab(x)+" "+grab_state(x))


df1.head()
```

```python
result2=df1.groupby('City').sum()
result2
```

```python
cites = [citii for citii,df11 in df1.groupby('City')]
plt.figure(figsize=(15,5))

plt.title("cities sales analysis of the",fontdict={"fontsize":20})
plt.xlabel("cities",fontdict={"fontsize":13})
plt.ylabel("sales cost in the us doller($)",fontdict={"fontsize":13})
plt.plot(cites,result2['Highest_Sales'],"g.-",label="city wise sales analysis")
plt.bar(cites,result2['Highest_Sales'],label="city wise sales analysis")
plt.legend()
plt.show()

```

```python
print("from the above bar graph we can tell that San Francisco CA has highest sales ")
```

### 4) what are the products most often sold together 

```python
df1.head()
```

```python
# so we want to find the item's sold together at ones .....

df2=df1[df1['Order ID'].duplicated(keep=False)]

df2.head(20)
```

```python
df2['Grouped']=df2.groupby('Order ID')['Product'].transform(lambda x : (",").join(x))
```

```python
 df2.head(10)
```

```python
# Removing the duplicate rows 

df2 = df2[['Order ID','Grouped']].drop_duplicates()
```

```python
df2.head(20)
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```
